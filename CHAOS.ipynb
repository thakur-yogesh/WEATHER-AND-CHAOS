```python
import sys
import traceback
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from random import randint
import random
from tkinter import *
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk 
from matplotlib.figure import Figure
from matplotlib import style
from mpl_toolkits.mplot3d import Axes3D 
import time
from tkinter import TclError
import matplotlib 
matplotlib.use("TkAgg") 
class LorenzAttractorRungeKutta():
    DT            = 1e-3     
    STEP          = 100   
    X_0, Y_0, Z_0 = 0,0,0 
    
    def __init__(self,*args, **kwargs):
        
        
        global user_r_entry
        global user_sigma_entry
        global user_b_entry
        global user_r_entry
        global user_sigma_entry
        global user_b_entry
        global user_x0_entry
        global user_y0_entry
        global user_z0_entry
        global graphframe
        global root
        global steps
        

            
        self.root = Tk()
        self.root.grid()
        self.root.title("Lorenz Simulation")
        self.root.geometry = ("1000x1000")
        self.root.resizable = (True, True)
        
        self.res = [[], [], []] 
        
        controlframe = LabelFrame(self.root, text = "Parameter Control")
        controlframe.grid(row = 1, column = 0) 
        
        #user_r_entry = DoubleVar()
        #r_entry = Entry(controlframe, 
                        #textvariable = user_r_entry).grid(row = 1, 
                                                   #column = 0)
        #self.root.r_label = Label(controlframe, 
                                  #text="r value", 
                                  #height = 1, 
                                  #width = 12).grid(row=2,
                                            #column=0, 
                                            #columnspan=1)
        steps = IntVar()
        steps_entry = Entry(controlframe, 
                        textvariable = steps).grid(row = 5, 
                                                   column = 1)
        self.root.r_label = Label(controlframe, 
                                  text="step value", 
                                  height = 1, 
                                  width = 12).grid(row=6,
                                            column=1, 
                                            columnspan=1)
        user_x0_entry = DoubleVar()
        X_0 = Entry(controlframe, 
                        textvariable = user_x0_entry).grid(row = 3, 
                                                   column = 0)
        self.root.r_label = Label(controlframe, 
                                  text="x0 value", 
                                  height = 1, 
                                  width = 12).grid(row=4,
                                            column=0, 
                                            columnspan=1)
        
        user_y0_entry = DoubleVar()
        Y_0 = Entry(controlframe, 
                        textvariable = user_y0_entry).grid(row = 3, 
                                                   column = 1)
        self.root.r_label = Label(controlframe, 
                                  text="y0 value", 
                                  height = 1, 
                                  width = 12).grid(row=4,
                                            column=1, 
                                            columnspan=1)
        user_z0_entry = DoubleVar()
        Z_0 = Entry(controlframe, 
                        textvariable = user_z0_entry).grid(row = 3, 
                                                   column = 2)
        self.root.r_label = Label(controlframe, 
                                  text="z0 value", 
                                  height = 1, 
                                  width = 12).grid(row=4,
                                            column=2, 
                                            columnspan=1)
        
        
        #user_sigma_entry = DoubleVar() 
        #sigma_entry = Entry(controlframe, 
                            #textvariable = user_sigma_entry).grid(row = 1, 
                                                           #column = 1)
        #self.root.sigma_label = Label(controlframe, 
                                     # text="sigma value", 
                                      #height = 1, 
                                     # width = 12).grid(row=2, 
                                                #column=1, 
                                               # columnspan=1)
        
        #user_b_entry = DoubleVar() 
        #b_entry = Entry(controlframe, 
                        #textvariable = user_b_entry).grid(row = 1, column = 2)
        #self.root.b_label = Label(controlframe, 
                                  #text="b value", 
                                  #height = 1, 
                                  #width = 12).grid(row=2, 
                                            #column=2, 
                                            #columnspan=1)
        
        self.plot_button1 = Button (self.root, 
                                    command = self.click1, 
                                    height = 2, 
                                    width = 8, 
                                    text = "Run").grid(row = 2, column = 0) 
        
        self.plot_button2 = Button (self.root, 
                                    command = self.Quit, 
                                    height = 2, 
                                    width = 8, 
                                    text = "Quit").grid(row = 4, column = 0)
        
        self.plot_button3 = Button (self.root, 
                            command = self.clear, 
                            height = 2, 
                            width = 8, 
                            text = "Clear").grid(row = 3, column = 0)
        
        
        graphframe = LabelFrame (self.root, text = "Graph")
        graphframe.grid(row = 0, column = 0)
        
        fig = Figure() 
        self.canvas = FigureCanvasTkAgg(fig, master = graphframe)
        
        self.root.mainloop()
            
    def click1(self):
        

        global r_info
        global sigma_info
        global b_info
        global user_r_entry
        global user_sigma_entry
        global user_b_entry
        global graphframe

        try:
            r_info = 28
            sigma_info = 10
            b_info = 2.66  
            
            self.X_0 = user_x0_entry.get()
            self.Y_0 = user_y0_entry.get()
            self.Z_0 = user_z0_entry.get()
            
            self.STEP = steps.get()
            
            print ("")
            print ("==============================================")
            print ("                 SYSTEM REPORT                ")              
            
            timeinit = time.process_time() 
                        
            print ("User entered: \n r     = %f \n sigma = %f \n b     = %f \n" 
                   % (r_info, sigma_info, b_info))
            
            print ("The randomised initial values are \n X0 = %f \n Y0 = %f \n Z0 = %f \n " 
                   %(self.X_0, self.Y_0, self.Z_0))
            
            print ("The number of steps are \n steps = %f \n  "
                   %(self.STEP))
            
            xyz = [self.X_0, self.Y_0, self.Z_0] 
            
            for _ in range(self.STEP): 
                k_0 = self.__lorenz(xyz)
                k_1 = self.__lorenz([x + k * self.DT / 2 for x, k in zip(xyz, k_0)])
                k_2 = self.__lorenz([x + k * self.DT / 2 for x, k in zip(xyz, k_1)])
                k_3 = self.__lorenz([x + k * self.DT for x, k in zip(xyz, k_2)])
                for i in range(3):
                    print(self.res[i])
                    xyz[i] += (k_0[i] + 2 * k_1[i] + 2 * k_2[i] + k_3[i]) \
                            * self.DT / 6.0
                    
                    self.res[i].append(xyz[i])
                    
            fig = Figure() 
            ax = Axes3D(fig)
            ax.set_xlabel("x")
            ax.set_ylabel("y")
            ax.set_zlabel("z")
            
            self.canvas = FigureCanvasTkAgg(fig, master = graphframe)
            ax.plot(self.res[0], self.res[1], self.res[2], color="blue", lw=1) 
    
            self.canvas.draw()
            self.canvas.get_tk_widget().grid(row = 0, column = 0)
            
            timeend = time.process_time() 
            timer =  timeend - timeinit 

        except tk.TclError:
            print (messagebox.showinfo("Invalid!", "Please Enter Valid Inputs (i.e. integers or floats)"))
            print ("")
            
    
    def clear(self): 
        self.root.destroy() 
        LorenzAttractorRungeKutta() 
        
    def __lorenz(self, xyz): 
        global r_info
        global sigma_info
        global b_info
        
        p = sigma_info 
        r = r_info
        b = b_info
        
        return [
                -p * xyz[0] + p * xyz[1], 
                -xyz[0] * xyz[2] + r * xyz[0] - xyz[1], 
                xyz[0] * xyz[1] - b * xyz[2]
                ] #return as a list

            
    def Quit(self):
        print ("Exiting Program...")
        self.root.quit() #Quit program


if __name__ == '__main__':
    try:
        LorenzAttractorRungeKutta() 
        
    except Exception as e:
        traceback.print_exc()
        sys.exit(1)


        
#sigma = 10 b = 8/3 r = 28
```
